```python
import numpy as np
from numpy.random import normal
#from numpy import random
import matplotlib.pyplot as plt
import statistics as st
import seaborn as sn
import random
```

```python
def montecarlo_1(n):
  mu = 5 # mean of the distribution
  sigma = 1 # variance of the distribution
  samples = normal(mu, sigma, size = (n)) # generating samples that have normal/Guassian distribution
  return samples
```

```python
x = montecarlo_1(1000)
plt.hist(x)
plt.show()
sn.distplot(x)
plt.show()
print("\n Mean of sample set is", st.mean(x))
print("\n Variance of sample set is % s", st.variance(x))
```

# B-day Paradox

```python
x = np.array(random.choices(range(10), k = 9))
print(x)
print(len(x) - len(np.unique(x)))
```

```python
# Selects bdays from 0 to 364 values for n number of people
def bday_paradox(n):
  n_data = np.array(random.choices(range(365), k = n))
  return n_data
```

```python
# Runs for sample number of iterations
sample = 1000
n = 23
shared_bday = 0
for i in range(sample):
  t = bday_paradox(n)
  if len(t) - len(set(t)) > 0:
    shared_bday += 1

print('Probabilty of more than 2 people sharing same birthday is :', shared_bday/sample)
```

```python
u = np.array(range(0,101,10))
print('n = ', u)
prob = []
for n in u:
  sample = 100
  shared_bday = 0
  for i in range(sample):
      t = bday_paradox(n)
      if len(t) - len(set(t)) > 0:
        shared_bday += 1
  prob.append(shared_bday/sample)
print(prob)
```

```python
# Plot number of people on x-axis
# prob of sharing the same bday date on y-axis
#
plt.plot(u, prob)
plt.grid(True)
plt.title('Birthday paradox')
plt.xlabel('Number of people')
plt.ylabel('Prob, of atleast 2 people sharing the same birthday date')
plt.show()
```
